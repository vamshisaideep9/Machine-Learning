```python
import pandas as pd
df = pd.read_csv("C:/Users/vamsh/OneDrive/Desktop/MACHINE LEARNING/L1 and L2 regularization/Melbourne_housing_FULL.csv")
df.head()
```

```python
df.nunique()
```

```python
df.shape
```

```python
df.drop(['Address','Date', 'Lattitude', 'Longtitude', 'YearBuilt', 'Postcode'], axis='columns', inplace=True)
df.head()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
cols_to_fill_zero = ['Propertycount','Distance','Bedroom2','Bathroom','Car']
df[cols_to_fill_zero] = df[cols_to_fill_zero].fillna(0)
```

```python
df.isna().sum()
```

```python
df['Landsize'] = df['Landsize'].fillna(df.Landsize.mean())
df['BuildingArea'] = df['BuildingArea'].fillna(df.BuildingArea.mean())
df.isna().sum()
```

```python
df.dropna(inplace=True)
df.isna().sum()
```

```python
df = pd.get_dummies(df, drop_first=True)
df.head()
```

```python
x = df.drop('Price', axis=1)
y = df['Price']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(x_train, y_train)
```

```python
reg.score(x_test,y_test)
```

```python
reg.score(x_train,y_train)
```

Above Model is Overfitting because the training gives more acc then testing

```python
#Lesso -> L1 regularization

from sklearn import linear_model
lesso_reg = linear_model.Lasso(alpha=50, max_iter=100, tol=0.1)
lesso_reg.fit(x_train,y_train)
```

```python
lesso_reg.score(x_train, y_train)
```

```python
lesso_reg.score(x_test, y_test)
```

```python
from sklearn.linear_model import Ridge

ridge_reg = Ridge(alpha=50, max_iter=100, tol=0.1)
ridge_reg.fit(x_train,y_train)
```

```python
ridge_reg.score(x_train,y_train)
```

```python
ridge_reg.score(x_test,y_test)
```
