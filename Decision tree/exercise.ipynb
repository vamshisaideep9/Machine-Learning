```python
import pandas as pd
```

```python
df = pd.read_csv("C:/Users/vamsh/OneDrive/Desktop/MACHINE LEARNING/Decision tree/titanic.csv")
df.head()
```

```python
drop_table = df.drop(['Ticket', 'Name', 'Cabin'], axis='columns')
drop_table
```

```python
inputs = drop_table.drop(['Survived','PassengerId','SibSp', 'Parch'], axis='columns')
inputs
```

```python
target = drop_table['Survived']
target
```

```python
from sklearn.preprocessing import LabelEncoder
le_sex = LabelEncoder()

inputs['sex_n'] = le_sex.fit_transform(inputs['Sex'])
inputs.head()
```

```python
x = inputs.drop(['Sex','Age_n','Embarked'], axis='columns')
x
```

```python
median_age = x['Age'].median()
median_age
```

```python
x['Age'].fillna(median_age, inplace=True)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x, target, test_size = 0.2, random_state=42)
```

```python
print(f"x_train shape: {x_train.shape}")
print(f"x_test shape: {x_test.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape: {y_test.shape}")
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test, y_test)
```

```python
model.predict(x_test)
```
